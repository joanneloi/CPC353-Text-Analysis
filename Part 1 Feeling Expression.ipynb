```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data for Anxiety File
collocates = ["scared", "tired", "anxious", "dying", "hopeless"]
likelihood = [122.980, 92.936, 71.922, 41.823, 37.234]

# Numeric positions for bars
x = np.arange(len(collocates)) * 0.2

# Create bar chart with closer bars
plt.figure(figsize=(8, 5))
plt.bar(x, likelihood, color='skyblue', edgecolor='black', width=0.1) 
plt.xticks(x, collocates)
plt.xlabel("Collocate", fontsize=14)
plt.ylabel("Likelihood", fontsize=14)
plt.title("Anxiety", fontsize=15)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
